```python
import keras
from keras.models import Sequential
from keras.datasets import mnist
from keras.layers import Dense, Activation, Conv2D, Flatten, Dropout, MaxPooling2D
from keras.utils import to_categorical
from keras_preprocessing.image import ImageDataGenerator
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
from keras.callbacks import History 
import tensorflow as tf
import numpy as np
```

```python
# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# reshape to have 1 color channel
x_train = np.reshape(x_train, (60000, 28, 28, 1))
x_test = np.reshape(x_test, (10000, 28, 28, 1))

#cast values to categorical data
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)


generator = ImageDataGenerator(
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=3,
        zoom_range=0.1,
        fill_mode="constant",
        cval=0,
        rotation_range=25,
        rescale=1./255)

generator.fit(x_train)

model = Sequential([
    Conv2D(64, kernel_size=3, activation='relu', input_shape=(28,28,1)),
    Conv2D(64, kernel_size=3, activation='relu'),
    MaxPooling2D(pool_size=2),
    #Dropout(0),
    Conv2D(128, kernel_size=3, activation='relu'),
    Conv2D(128, kernel_size=3, activation='relu'),
    MaxPooling2D(pool_size=2),
    Flatten(),
    Dense(256, activation='relu'),
    #Dropout(0.4),
    Dense(10, activation='softmax')
])

es = EarlyStopping(monitor='val_accuracy', mode='max', verbose=1, patience=10, restore_best_weights=True)
history = History()

model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])

model.fit_generator(generator.flow(x_train, y_train, batch_size=128),
          validation_data=(x_test, y_test),
          epochs=50,
          verbose=1,
          callbacks=[es, history]  
          )
```

```python
model.save("cnn1.h5")
```
